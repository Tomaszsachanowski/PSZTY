# Import pobrzebnych funkcji do irisow

```python
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA
```

# Stworzenie zbiru irisow

```python
iris = datasets.load_iris()
```

# Bibiloteka do rysowania

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook

# import some data to play with
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(3, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())



X = iris.data[:, 2:]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')


plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())


fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:50, 0], X_reduced[:50, 1], X_reduced[:50, 2], c='g',
            s=10)
ax.scatter(X_reduced[50:100, 0], X_reduced[50:100, 1], X_reduced[50:100, 2], c='y',
            s=10)
ax.scatter(X_reduced[100:, 0], X_reduced[100:, 1], X_reduced[100:, 2], c='r',
            s=10)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```

# WyÅ›wietlenie naszych pÅ‚aszczyn

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D





xx, yy = np.meshgrid(range(-3,3,1), range(-3,3,1))
plt3d = plt.figure().gca(projection='3d')

# calculate corresponding z
for os in pop.populacja_P:
    d = os.wektor_wspol_w[0]
    a = os.wektor_wspol_w[1]
    b = os.wektor_wspol_w[2]
    c = os.wektor_wspol_w[3]
    z = (-a * xx - b * yy - d) * 1. / c

    # plot the surface
    # plot the surface
    plt3d.plot_surface(xx, yy, z, alpha=0.2)

for point in X:
        if point[-1] == 1:
            plt3d.scatter(point[0], point[1], point[2], c='b', s=10)
        else:
            plt3d.scatter(point[0], point[1], point[2], c='g', s=10)
            
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```
